```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=None, hspace=None)

st = np.vstack([rec_hst[...,0], rec_hst_2[...,0]])

ims = []
for i in range(128):
    im = plt.imshow(st[i,16:-16,16:-16], animated=True,cmap='gray',vmin=0,vmax=0.17)
    plt.axis('off')
    ims.append([im])
fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=None, hspace=None)

ani = animation.ArtistAnimation(fig, ims, interval=200, blit=True,
                                repeat_delay=1000)

ani.save('rec_samples.mp4',dpi=150)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=None, hspace=None)


st = np.vstack([rec_lsst, rec_lsst_2])

ims = []
for i in range(128):
    im = plt.imshow(st[i,4:-4,4:-4,0] + 0.002*randn(24,24),vmin=-0.006,vmax=0.035, animated=True,cmap='gray')
    plt.axis('off')
    ims.append([im])
    
fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=None, hspace=None)

ani = animation.ArtistAnimation(fig, ims, interval=200, blit=True,
                                repeat_delay=1000)

ani.save('rec_lsst_noise.mp4',dpi=150)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=None, hspace=None)

st = np.vstack([rec_lsst, rec_lsst_2])

ims = []
for i in range(128):
    im = plt.imshow(obs_lsst[0,4:-4,4:-4,0] - st[i,4:-4,4:-4,0],vmin=-0.006,vmax=0.035, animated=True,cmap='gray')
    plt.axis('off')
    ims.append([im])
    
fig.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=None, hspace=None)
ani = animation.ArtistAnimation(fig, ims, interval=200, blit=True,
                                repeat_delay=1000)

ani.save('res_lsst.mp4',dpi=150)

plt.show()
```
